```python
import csv

# Sample data (you can replace this with your own data)
data = [
    {"country": "USA", "life_expectancy": 78.6, "gdp_per_capita": 63479},
    {"country": "China", "life_expectancy": 77.3, "gdp_per_capita": 10411},
    {"country": "Japan", "life_expectancy": 84.3, "gdp_per_capita": 43523},
    {"country": "India", "life_expectancy": 70.8, "gdp_per_capita": 2264},
    {"country": "Germany", "life_expectancy": 81.4, "gdp_per_capita": 49348},
]

# Create a CSV file
with open("world_data.csv", "w", newline="") as csvfile:
    fieldnames = ["country", "life_expectancy", "gdp_per_capita"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(data)
```

```python
import seaborn as sns
import pandas as pd
```

```python

# Read the CSV data into a pandas DataFrame
df = pd.read_csv("world_data.csv")

# Create a scatter plot using Seaborn
sns.set_style("whitegrid")
sns.scatterplot(x="gdp_per_capita", y="life_expectancy", hue="country", data=df)

# Add a title and axis labels
sns.plt.title("Life Expectancy vs. GDP per Capita")
sns.plt.xlabel("GDP per Capita (USD)")
sns.plt.ylabel("Life Expectancy (Years)")

# Show the plot
sns.plt.show()
```
