```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Basic Plotting

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

# font property
fontdict={'font':'Comic Sans MS', 'fontsize':'10'}

# plotting with parameters colour, linewidth, marker, label for legend
plt.plot(x,y, color='green', linewidth='2', marker='o', label='X')

plt.title('My First Graph',fontdict=fontdict )

plt.xlabel('X Axis',fontdict=fontdict)
plt.ylabel('Y Axis',fontdict=fontdict)

plt.xticks([1,2,3,4,5])
plt.yticks([1,2,3,4,5])

plt.legend()

plt.show()
```
